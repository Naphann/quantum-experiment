```python
from qiskit import *
from qiskit.compiler import transpile
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor
IBMQ.load_accounts()
%matplotlib inline
```

```python
a = QuantumRegister(3, 'a')
b = QuantumRegister(3, 'b')
c = QuantumRegister(2, 'c')
t = QuantumRegister(2, 't')
r = QuantumRegister(1, 'r')
cr = ClassicalRegister(1, 'cr')
cmp = QuantumCircuit(a, b, c, t, r, cr, name='cmp')

cmp.h(a)
cmp.x(b[0])
cmp.barrier()

cmp.ccx(a[0], b[0], c[0]) # 1
cmp.ccx(a[1], b[1], c[1]) # 2
cmp.x(b[0])               # 3
cmp.ccx(a[0], b[0], r[0]) # r
cmp.x(a[0])               # 4
cmp.ccx(a[0], b[0], c[0]) # 5
cmp.x(b[1])               # 6
cmp.ccx(a[1], b[1], t[0]) # 7
cmp.ccx(c[0], t[0], r[0]) # r
cmp.ccx(a[1], b[1], t[0]) # -7
cmp.x(a[1])               # 8
cmp.ccx(a[1], b[1], c[1]) # 9
cmp.ccx(c[0], c[1], t[1]) # 10
cmp.x(b[2])               # 11
cmp.ccx(a[2], b[2], t[0]) # 12
cmp.ccx(t[0], t[1], r)    # r
cmp.ccx(a[2], b[2], t[0]) # -12
cmp.x(b[2])               # -11
cmp.ccx(c[0], c[1], t[1]) # -10
cmp.ccx(a[1], b[1], c[1]) # -9
cmp.x(a[1])               # -8
cmp.x(b[1])               # -6
cmp.ccx(a[0], b[0], c[0]) # -5
cmp.x(a[0])               # -4
cmp.x(b[0])               # -3
cmp.ccx(a[1], b[1], c[1]) # -2
cmp.ccx(a[0], b[0], c[0]) # -1

cmp.barrier()
cmp.measure(r, cr)

print('gates = ', cmp.count_ops())
print('depth = ', cmp.depth())
cmp.draw(output='latex', scale=0.5)
```

```python
device = IBMQ.get_backend('ibmq_16_melbourne')
trans_cmp = transpile(cmp, device)

print('gates = ', trans_cmp.count_ops())
print('depth = ', trans_cmp.depth())
# trans_cmp.draw(output='mpl')
```

```python
job = execute([cmp, trans_cmp], device)
job_monitor(job)
res = job.result()
```

```python
# now try with simulator to see the ideal results
sim_backend = IBMQ.get_banckend('ibm_qasm_simulator')
sim_job = execute([cmp], sim_backend)
job_monitor(sim_job)
sim_res = sim_job.result()
```

```python
plot_histogram([res.get_counts(0), res.get_counts(1), sim_res.get_counts()],
               figsize=(15,5),
              legend=['default', 'transpiled', 'ideal'])
```
