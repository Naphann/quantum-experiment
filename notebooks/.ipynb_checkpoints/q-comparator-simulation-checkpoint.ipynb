```python
def qcomp_3b(qs):
    '''
    a  : 3-bit qreg
    b  : 3-bit qreg
    r  : 1-bit result 1 if a > b otherwise 0
    anc: 4-bit qreg 
    return new circuit
    '''
    # TODO: checking for number of bits
    cmp = QuantumCircuit(qs)
    cmp.ccx(qs[0], qs[3], qs[6])  # 1
    cmp.ccx(qs[1], qs[4], qs[7])  # 2
    cmp.x(qs[3])                  # 3
    cmp.ccx(qs[0], qs[3], qs[10]) # r
    cmp.x(qs[0])                  # 4
    cmp.ccx(qs[0], qs[3], qs[6])  # 5
    cmp.x(qs[4])                  # 6
    cmp.ccx(qs[1], qs[4], qs[8])  # 7
    cmp.ccx(qs[6], qs[8], qs[10]) # r
    cmp.ccx(qs[1], qs[4], qs[8])  # -7
    cmp.x(qs[1])                  # 8
    cmp.ccx(qs[1], qs[4], qs[7])  # 9
    cmp.ccx(qs[6], qs[7], qs[9])  # 10
    cmp.x(qs[5])                  # 11
    cmp.ccx(qs[2], qs[5], qs[8])  # 12
    cmp.ccx(qs[8], qs[9], qs[10]) # r
    cmp.ccx(qs[2], qs[5], qs[8])  # -12
    cmp.x(qs[5])                  # -11
    cmp.ccx(qs[6], qs[7], qs[9])  # -10
    cmp.ccx(qs[1], qs[4], qs[7])  # -9
    cmp.x(qs[1])                  # -8
    cmp.x(qs[4])                  # -6
    cmp.ccx(qs[0], qs[3], qs[6])  # -5
    cmp.x(qs[0])                  # -4
    cmp.x(qs[3])                  # -3
    cmp.ccx(qs[1], qs[4], qs[7])  # -2
    cmp.ccx(qs[0], qs[3], qs[6])  # -1
    return cmp

def qcomp_3b_2(a, b, r, anc):
    '''
    a  : 3-bit qreg
    b  : 3-bit qreg
    r  : 1-bit result 1 if a > b otherwise 0
    anc: 4-bit qreg 
    return new circuit
    '''
    # TODO: checking for number of bits
    cmp = QuantumCircuit(a, b, r, anc)
    cmp.ccx(a[0], b[0], anc[0]) # 1
    cmp.ccx(a[1], b[1], anc[1]) # 2
    cmp.x(b[0])               # 3
    cmp.ccx(a[0], b[0], r[0]) # r
    cmp.x(a[0])               # 4
    cmp.ccx(a[0], b[0], anc[0]) # 5
    cmp.x(b[1])               # 6
    cmp.ccx(a[1], b[1], anc[2]) # 7
    cmp.ccx(anc[0], anc[2], r[0]) # r
    cmp.ccx(a[1], b[1], anc[2]) # -7
    cmp.x(a[1])               # 8
    cmp.ccx(a[1], b[1], anc[1]) # 9
    cmp.ccx(anc[0], anc[1], anc[3]) # 10
    cmp.x(b[2])               # 11
    cmp.ccx(a[2], b[2], anc[2]) # 12
    cmp.ccx(anc[2], anc[3], r)    # r
    cmp.ccx(a[2], b[2], anc[2]) # -12
    cmp.x(b[2])               # -11
    cmp.ccx(anc[0], anc[1], anc[3]) # -10
    cmp.ccx(a[1], b[1], anc[1]) # -9
    cmp.x(a[1])               # -8
    cmp.x(b[1])               # -6
    cmp.ccx(a[0], b[0], anc[0]) # -5
    cmp.x(a[0])               # -4
    cmp.x(b[0])               # -3
    cmp.ccx(a[1], b[1], anc[1]) # -2
    cmp.ccx(a[0], b[0], anc[0]) # -1
    return cmp
```

```python
# creating the circuit

# a = QuantumRegister(3, 'a')
# b = QuantumRegister(3, 'b')
# anc = QuantumRegister(4, 'c')
# r = QuantumRegister(1, 'r')
qs = QuantumRegister(11, 'qs')
qc = QuantumCircuit(qs)

qcmp = qcomp_3b(qs)

qc.h(qs[0])
qc.x(qs[3])
qc.barrier()

# qcmp2 = qcomp_3b(a, b, anc)

qctmp = qc + qcmp

qctmp.draw(output='latex', scale=0.5)
```

```python
# Create the expected density matrix
# Since for now we only care about register 'a'
a = QuantumRegister(3, 'a')
aqc = QuantumCircuit(a)
aqc.h(a[0])
print(aqc)

job = qiskit.execute(aqc, Aer.get_backend('statevector_simulator'))
psi_aqc = job.result().get_statevector(aqc)
print(psi_aqc)
```

```python
# since we already have the actual circuit to use
# 'qctmp' we can run it
t = time.time()
qst_aqc = state_tomography_circuits(qctmp, [qs[0], qs[1], qs[2]])
job = qiskit.execute(qst_aqc, Aer.get_backend('qasm_simulator'), shots=5000)
print('Execution - time:', time.time() - t)

t = time.time()
tomo_aqc = StateTomographyFitter(job.result(), qst_aqc)
print('Fitter    - time taken:', time.time() - t)
```

```python
# perform the tomography fit
# which outputs a density matrix
t = time.time()
rho_aqc = tomo_aqc.fit()
F_aqc = state_fidelity(psi_aqc, rho_aqc)
print('Fit - time taken:', time.time() - t)
print('Fit Fidelity = ', F_aqc)
```

### Repeat the Experiment with Measurement Noise

```python
# add measurement noise
noise_model = noise.NoiseModel()
for qi in range(11):
    read_err = noise.errors.readout_error.ReadoutError([[0.8, 0.2], [0.2, 0.8]])
    noise_model.add_readout_error(read_err, [qi])
    
# generate the calibration circuits
meas_calibs, state_labels = mc.complete_meas_cal(qubit_list=[0,1,2])

backend = Aer.get_backend('qasm_simulator')
job_cal = qiskit.execute(meas_calibs, backend=backend, shots=15000, noise_model=noise_model)
job_tomo = qiskit.execute(qst_aqc, backend=backend, shots=15000, noise_model=noise_model)

meas_fitter = mc.CompleteMeasFitter(job_cal.result(), state_labels)

tomo_aqc = StateTomographyFitter(job_tomo.result(), qst_aqc)

# no correction
rho_aqc = tomo_aqc.fit()
F_aqc = state_fidelity(psi_aqc, rho_aqc)
print('Fit Fidelity (no correction) =', F_aqc) 
```

```python
# correct data
correct_tomo_results = meas_fitter.filter.apply(job_tomo.result(), method='least_squares')
tomo_aqc = StateTomographyFitter(correct_tomo_results, qst_aqc)
rho_aqc = tomo_aqc.fit()
F_aqc = state_fidelity(psi_aqc, rho_aqc)
print('Fit Fidelity (w/ correction) =', F_aqc) 
```

## Generate and fitting random states

```python
def random_cmp_tomo(shots=10):
    
    def rand_angles():
        return tuple(2 * np.pi * np.random.random(3) - np.pi)
    
    def rand_int_reg():
        return True
    
    q = QuantumRegister(11)
    _circ = QuantumCircuit(q)
    
    _ideal_circ = QuantumCircuit(3)
    
    cmp = qcomp_3b(q)
    for j in range(3):
        args = rand_angles()
        _circ.u3(*args, q[j])
        _ideal_circ.u3(*args, j)
    for j in range(3,6):
        if np.random.random() > 0.5:
            _circ.x(q[j])
    circ = _circ + cmp
    job = qiskit.execute(_ideal_circ, Aer.get_backend('statevector_simulator'), shots=shots)
    psi_rand = job.result().get_statevector(_ideal_circ)
    
    qst_circs = state_tomography_circuits(circ, [q[0], q[1], q[2]])
    job = qiskit.execute(qst_circs, Aer.get_backend('qasm_simulator'), shots=shots)
    tomo_data = StateTomographyFitter(job.result(), qst_circs)
    rho_cvx = tomo_data.fit(method='cvx')
    rho_lstsq = tomo_data.fit(method='lstsq')
    
    print('F fit (CVX)   = ', state_fidelity(psi_rand, rho_cvx))
    print('F fit (LSTSQ) = ', state_fidelity(psi_rand, rho_lstsq))
```

```python
for j in range(5):
    print('Random comparator: set {}'.format(j+1))
    random_cmp_tomo(20000)
```

### Algorithm Outline

1. initialize 0 -> a, random -> b
2. setup output by H(X(output))
3. H(a)
4. Uf(a, b)
5. grover diffusion (a)
6. repeat 4-5 by a fixed number k (maybe randomly chosen)
7. measure a 

```python
def n_control_not_simple(qc, reg, output, anc):
    _qc = qc
    
    if (len(reg) == 2):
        _qc.ccx(reg[0], reg[1], output)
        return _qc
    
    _qc.ccx(reg[0], reg[1], anc[0])
    
    for i in range(2, len(reg)-1):
        _qc.ccx(reg[i], anc[i-2], anc[i-1])
    
    _qc.ccx(reg[-1], anc[len(reg)-3], output)
    
    for i in reversed(range(2, len(reg)-1)):
        _qc.ccx(reg[i], anc[i-2], anc[i-1])
                      
    _qc.ccx(reg[0], reg[1], anc[0])
        
    
    return _qc
    
def n_control_not_v_gate(qc, controls, target, anc):
    _qc = qc
    _qs = [controls[i] for i in len(controls)]
    
    def _cnot_recurse(qc, qs, tgt):
        if len(qs) == 2:
            qc.ccx(qs[0], qs[1], tgt)
        elif len(qs) == 1:
            qc.cx(qs[0], tgt)
        else:
            qc.cv(qs[-1], tgt)
            _cnot_recurse(qc, qs[:-1], qs[-1])
            qc.cvdg(qs[-1], tgt)

            
            

def U_f(qc, a, b, anc, output):
    _qc = U_id(qc)
    _qcomp = qcomp_3b_2(a, b, output, anc)
    
def init_grover(qc, a, b, anc, output):
    _qc = qc
    _qc.x(output)
    _qc.h(output)
    _qc.h(a)
    
def grover_diffusion(qc, a, b, anc, output):
    _qc = qc
    _qc.h(a)
    _qc.x(a)
    _qc.h(a[0])
    _qc = n_control_not(_qc, a, anc)
    _qc.h(a[0])
    _qc.x(a)
    _qc.h(a)
    return _qc
```
