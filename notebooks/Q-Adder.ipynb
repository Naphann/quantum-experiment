# Adder Circuit

```python
# needed for functions
import numpy as np
import time
from copy import deepcopy

# import Qiskit classes
import qiskit
from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister, Aer
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor

%matplotlib inline
```

```python
# Adder 
def adder(n, a, b, anc, result):
    # check ancilla register count
    if len(anc) < 1:
        # error 
        print('ancilla qubit not enough')
        return
    # check a and b length
    if len(a) < n or len(b) < n:
        print('a or b has length less than specify n:', n)
        return
    # check result register length
    if len(result) < n:
        print('register has length than specify n:', n)
        return
    
    # anc[0] is for overflow from last bit
    
    qc = QuantumCircuit(a, b, anc, result)
    
    qc.ccx(a[0], b[0], anc[0]) # -- set anc[0]
    qc.cx(a[0], result[0])
    qc.cx(b[0], result[0])
    
    for j in range(1, n):
        qc.cx(a[j], result[j])
        qc.cx(b[j], result[j])
        qc.cx(anc[0], result[j])
        
        # clear anc[0] i.e. carry bit from last bit
        qc.ccx(a[j-1], b[j-1], anc[0])
        qc.x(result[j-1]) # -- X result
        qc.ccx(a[j-1], result[j-1], anc[0])
        qc.ccx(b[j-1], result[j-1], anc[0])
        qc.x(result[j-1]) # -- X result
        
        
        # set anc[0] if there is carry from this bit
        qc.ccx(a[j], b[j], anc[0])
        qc.x(result[j]) # -- X result
        qc.ccx(a[j], result[j], anc[0])
        qc.ccx(b[j], result[j], anc[0])
        qc.x(result[j]) # -- X result
    
    # clear anc[0] from last bit
    qc.ccx(a[n-1], b[n-1], anc[0])
    qc.x(result[n-1]) # -- X result
    qc.ccx(a[n-1], result[n-1], anc[0])
    qc.ccx(b[n-1], result[n-1], anc[0])
    qc.x(result[n-1]) # -- X result
    
    return qc
```

### Now let's see how our adder circuit looks like

```python
a = QuantumRegister(4, 'a')
b = QuantumRegister(4, 'b')
anc = QuantumRegister(1, 'anc')
result = QuantumRegister(4, 'result')
qc = adder(4, a, b, anc, result)
qc.draw(output='latex')
```

```python
def set_reg(i, qc, reg):
    # TODO check length
    cur = 1
    j = 0
    while (cur <= i):
        mark = cur & i
        if mark > 0:
            qc.x(reg[j])
        cur <<= 1
        j += 1
```

```python
def init_w_int(n, an, bn):
    a = QuantumRegister(n, 'a')
    b = QuantumRegister(n, 'b')
    anc = QuantumRegister(1, 'anc')
    result = QuantumRegister(n, 'result')
    qc = QuantumCircuit(a, b, anc, result)
    set_reg(an, qc, a)
    set_reg(bn, qc, b)
    qc.barrier()
    return qc
```

```python
def check_adder_result(n, a, b, res):
    # TODO check for case where a + b > + 1 << n
    tmp = a + b
    if tmp >= 1<<n:
        tmp -= 1<<n
    return tmp == res
```

```python
def test_adder(n, an, bn):
    a = QuantumRegister(n, 'a')
    b = QuantumRegister(n, 'b')
    anc = QuantumRegister(1, 'anc')
    result = QuantumRegister(n, 'result')
    cs = ClassicalRegister(n, 'cs')
    
    _measure = QuantumCircuit(a, b, anc, result, cs)
    _measure.measure(result, cs)
    
    _adder = adder(n, a, b, anc, result)
    _adder.barrier()
    
    _qc = init_w_int(n, an, bn)
    
    qc = _qc + _adder + _measure
    
    backend = Aer.get_backend('qasm_simulator')
    job = qiskit.execute(qc, backend, shots=100)
    sum_result = next(iter(job.result().get_counts(qc).items()))
    if sum_result[1] == 100 and check_adder_result(n, an, bn, int(sum_result[0], 2)):
        print('Success:', an, '+', bn, '=', int(sum_result[0], 2))
    else:
        print('Error adding:', an, 'and', bn, 'expected', an + bn, 'got', int(sum_result[0], 2))
```

```python
# testing adder 
for i in range(16):
    for j in range(16):
        test_adder(4, i, j)
```

```python
def test_side_effects(n, a, b, anc, an, bn, checkReg, expected):
    result = QuantumRegister(n, 'result')
    cs = ClassicalRegister(len(checkReg), 'cs')
    
    _measure = QuantumCircuit(a, b, anc, result, cs)
    _measure.measure(checkReg, cs)
    
    _adder = adder(n, a, b, anc, result)
    _adder.barrier()
    
    _qc = init_w_int(n, an, bn)
    
#     print('_qc')
#     print(_qc.qregs)
#     print(_qc.cregs)
    
#     print('_adder')
#     print(_adder.qregs)
#     print(_adder.cregs)
    
#     print('_measure')
#     print(_measure.qregs)
#     print(_measure.cregs)
    
    qc = _qc + _adder
    qc = qc + _measure
    
    backend = Aer.get_backend('qasm_simulator')
    job = qiskit.execute(qc, backend, shots=100)
    a_result = next(iter(job.result().get_counts(qc).items()))
    if a_result[1] == 100 and expected == int(a_result[0], 2):
        print('No side effect on', checkReg.name ,'for:', an, '+', bn)
    else:
        print('Error on:', an, 'and', bn, 'expected', an, 'got', int(a_result[0], 2))
```
