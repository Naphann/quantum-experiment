```python
import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import execute
```

# Circuit Basics

```python
# create a quantum register with 3 qubits.
q = QuantumRegister(3, 'q')

# create a quantum circuit acting on q register.
circuit = QuantumCircuit(q)
```

Make GHZ state

```python
# add a H gate on qubit 0, putting this qubit in superposition.
circuit.h(q[0])

# add a CX gate on control qubit 0 and target qubit 1
circuit.cx(q[0], q[1])

# add qubit 2 in ghz state
circuit.cx(q[0], q[2])
```

# Visualizing Circuit

you can visualize circuit using `QuantumCircuit.draw()` or `print(QuantumCircuit)`

```python
circuit.draw()
```

```python
print(circuit)
```

you can also draw the circuit using LaTex by `QuantumCircuit.draw(output='latex')` or using matplotlib by passing `output='mpl'`

```python
circuit.draw(output='latex')
```

```python
circuit.draw(output='mpl')
```

# Simulating circuits using Qiskit Aer

```python
# Import Aer
from qiskit import BasicAer

# run the quantum circuit on a statevector simulator backend
backend = BasicAer.get_backend('statevector_simulator')
```

```python
# create a quantum program for execution
job = execute(circuit, backend)
```

```python
result = job.result()
```

```python
outputstate = result.get_statevector(circuit, decimals=3)
print(outputstate)
```

Qiskit also provides a visualization toolbox to allow you to view these results.

```python
from qiskit.tools.visualization import plot_state_city
plot_state_city(outputstate)
```

## Unitary Backend

`unitary_simulator` calculates $2^n \times 2^n$ matrix representing the gates in the quantum circuit.

```python
# run the quantum circuit on a unitary simulator backend
backend = BasicAer.get_backend('unitary_simulator')
job = execute(circuit, backend)
result = job.result()

# show results
print(result.get_unitary(circuit, decimals=3))
```

## OpenQASM backend

this can be used with measurements

```python
# create a classical register with 3 bits.
c = ClassicalRegister(3, 'c')
# create a Quantum circuit
meas = QuantumCircuit(q, c)
meas.barrier(q)
# map the quantum measurement to the classical bits
meas.measure(q, c)

# the qiskit circuit object supports composition using
# the addition operator.
qc = circuit + meas

# drawing the circuit
qc.draw(output='latex')
```

To simulate the circuit, we use the `qasm_simulator` in Qiskit Aer.

```python
# use Aer's qasm_simulator
backend_sim = BasicAer.get_backend('qasm_simulator')

# execute the circuit on the qasm simulator.
# shots is set to be 1024 by default
job_sim = execute(qc, backend_sim, shots=1024)

# grab the result
result_sim = job_sim.result()
```

to access counts from the result object use `get_counts(circuit)`

```python
counts = result_sim.get_counts(qc)
print(counts)
```

you can use `plot_histogram` to view the outcomes

```python
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

## Running circuits using IBMQ provider

```python
from qiskit import IBMQ
```

```python
IBMQ.load_accounts()
```

```python
print('available backends:')
IBMQ.backends()
```

we will choose the least busy device (which has at least 3 qubits) to run our program.

```python
from qiskit.providers.ibmq import least_busy

large_enough_devices = IBMQ.backends(filters=lambda x: x.configuration().n_qubits > 4 and not x.configuration().simulator)
backend = least_busy(large_enough_devices)
print('the least busy backend is {} which has {} qubits'.format(backend.name(), backend.configuration().n_qubits))
```

to run the circuit on the backend, we need to specify the number of shots and the number of credit we are willing to spend to run the current circuit.

```python
from qiskit.tools.monitor import job_monitor
shots = 1024 # maximum is 8192
max_credits = 3

job_exp = execute(qc, backend=backend, shots=shots, max_credits=max_credits)
job_monitor(job_exp)
```

```python
result_exp = job_exp.result()
```

```python
counts_exp = result_exp.get_counts(qc)
plot_histogram([counts_exp, counts])
```

## simulating circuits using HPC simulator

```python
backend = IBMQ.get_backend('ibmq_qasm_simulator', hub=None)
```

```python
shots = 1024
max_credits = 3

job_hpc = execute(qc, backend=backend, shots=shots, max_credits=max_credits)
```

```python
result_hpc = job_hpc.result()
```

```python
counts_hpc = result_hpc.get_counts(qc)
plot_histogram(counts_hpc)
```

## Retrieving a previously ran job

```python
jobID = job_exp.job_id()

print('JOB ID: {}'.format(jobID))
```

Given job id, we can later reconstructed job object from the backend using the `retrieve_job`

```python
job_get = backend.retrieve_job(jobID)
```

```python
job_get.result().get_counts(qc)
```
